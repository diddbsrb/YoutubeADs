```python
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/동국대학교/nlp스터디/유튜브_자막_분석/종합본_추가.csv", encoding='cp949')
```

```python
df
```

```python
# 문장 리스트 예시
sentences = df['요약문'].tolist()
```

```python
# 형태소 분석기를 사용하여 텍스트 전처리
okt = Okt()
def preprocess(text):
    tokens = okt.nouns(text)  # 명사만 추출
    return ' '.join(tokens)

# 문장 리스트 전처리
processed_sentences = [preprocess(sentence) for sentence in sentences]
```

```python
# TF-IDF 벡터화
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(processed_sentences)
feature_names = vectorizer.get_feature_names_out()

# 각 문장에서 TF-IDF 점수가 높은 상위 키워드 추출
def extract_keywords(tfidf_matrix, feature_names, top_n=5):
    keywords = []
    for doc in tfidf_matrix:
        # TF-IDF 점수가 높은 순으로 정렬
        sorted_indices = doc.toarray()[0].argsort()[::-1]
        top_keywords = [feature_names[idx] for idx in sorted_indices[:top_n]]
        keywords.append(top_keywords)
    return keywords

# 키워드 추출
keywords = extract_keywords(tfidf_matrix, feature_names)

# 결과 출력
for i, kw in enumerate(keywords):
    print(f"문장 {i+1}의 키워드: {', '.join(kw)}")
```
