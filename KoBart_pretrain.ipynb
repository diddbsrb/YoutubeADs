```python
import pandas as pd
from datasets import Dataset
from transformers import PreTrainedTokenizerFast
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
data = pd.read_csv("C:/동국대학교/nlp스터디/유튜브_자막_분석/code/파일들/종합본.csv" , encoding = 'cp949')
data
```

```python
data['스크립트_길이'] = data['스크립트'].apply(len)
data['요약문_길이'] = data['요약문'].apply(len)

# 최대 길이 찾기
최대_스크립트_길이 = data['스크립트_길이'].max()
최대_요약문_길이 = data['요약문_길이'].max()
최소_스크립트_길이 = data['스크립트_길이'].min()
최소_요약문_길이 = data['요약문_길이'].min()


print(f"스크립트의 최대 길이: {최대_스크립트_길이}")
print(f"요약문의 최대 길이: {최대_요약문_길이}")
print(f"스크립트의 최소 길이: {최소_스크립트_길이}")
print(f"요약문의 최소 길이: {최소_요약문_길이}")
```

```python
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer
from datasets import Dataset
import pandas as pd
```

```python
# 데이터셋 생성
dataset = Dataset.from_pandas(data)
```

```python
# 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained("psyche/KoT5-summarization")
```

```python
# 데이터 전처리 함수
def preprocess_function(examples):
    inputs = [doc for doc in examples['스크립트']]
    model_inputs = tokenizer(inputs, max_length=857, truncation=True, padding="max_length")

    # Setup the tokenizer for targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(examples['요약문'], max_length=95, truncation=True, padding="max_length")

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
# 데이터셋 전처리
tokenized_datasets = dataset.map(preprocess_function, batched=True)
```

```python
# 모델 로드
model = AutoModelForSeq2SeqLM.from_pretrained("psyche/KoT5-summarization")
```

```python
# 트레이닝 아규먼트 설정
training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    evaluation_strategy="no",                # 평가 전략을 "no"로 설정하여 평가를 하지 않음
    learning_rate=0.00001,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=5,
    predict_with_generate=True,
    logging_dir='./logs',                    # 로그 파일을 저장할 디렉토리
    logging_steps=10,                        # 로그를 기록할 학습 스텝 간격
    logging_strategy="steps",                # 로그를 기록할 전략
)
```

```python
# 트레이너 설정
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets,        # 모든 데이터를 학습 데이터셋으로 사용
    tokenizer=tokenizer
)
```

```python
# 모델 학습
trainer.train()
```

```python
new_scripts = [
   '평생이라는 말에는 평생 앞으로도라는 믿음이 담겨있다. 언제까지나 지금처럼 깨끗하고 맑을거란 믿음 제주 삼다수는 늘 그래왔으니까 한결같으니까 평생의 물로 삼다 우리가 믿는 물 제주 삼다수.'
]
```

```python
for script in new_scripts:
    # 입력 데이터 토큰화
    inputs = tokenizer(script, return_tensors="pt", padding="max_length", truncation=True, max_length=1000)

    # 요약 생성
    summary_ids = model.generate(
        input_ids=inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        bos_token_id=model.config.bos_token_id,
        eos_token_id=model.config.eos_token_id,
        length_penalty=1.0,
        max_length=100,
        min_length=30,
        num_beams=6,
        repetition_penalty=1.5,
        no_repeat_ngram_size=15,
    )

    # 요약 결과 디코딩
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # 결과 출력
    print(f"원본 스크립트: {script}")
    print(f"요약문: {summary}\n")
```
