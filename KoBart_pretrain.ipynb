```python
import pandas as pd
from datasets import Dataset
from transformers import PreTrainedTokenizerFast
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer
from datasets import Dataset
import pandas as pd
```

```python
#  
dataset = Dataset.from_pandas(data)
```

```python
#  
tokenizer = AutoTokenizer.from_pretrained("psyche/KoT5-summarization")
```

```python
#   
def preprocess_function(examples):
    inputs = [doc for doc in examples['']]
    model_inputs = tokenizer(inputs, max_length=857, truncation=True, padding="max_length")

    # Setup the tokenizer for targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(examples[''], max_length=95, truncation=True, padding="max_length")

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
#  
tokenized_datasets = dataset.map(preprocess_function, batched=True)
```

```python
#  
model = AutoModelForSeq2SeqLM.from_pretrained("psyche/KoT5-summarization")
```

```python
#   
training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    evaluation_strategy="no",                #   "no"    
    learning_rate=0.00001,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=5,
    predict_with_generate=True,
    logging_dir='./logs',                    #    
    logging_steps=10,                        #     
    logging_strategy="steps",                #   
)
```

```python
#  
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets,        #     
    tokenizer=tokenizer
)
```

```python
#  
trainer.train()
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Fine-tuning   
model_path = "path_to_save_your_fine_tuned_model
tokenizer = AutoTokenizer.from_pretrained("psyche/KoT5-summarization")
model = AutoModelForSeq2SeqLM.from_pretrained("psyche/KoT5-summarization")
```

```python
new_scripts = [
   '     .                  .'
]
```

```python
for script in new_scripts:
    #   
    inputs = tokenizer(script, return_tensors="pt", padding="max_length", truncation=True, max_length=1000)

    #  
    summary_ids = model.generate(
        input_ids=inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        bos_token_id=model.config.bos_token_id,
        eos_token_id=model.config.eos_token_id,
        length_penalty=1.0,
        max_length=100,
        min_length=30,
        num_beams=6,
        repetition_penalty=1.5,
        no_repeat_ngram_size=15,
    )

    #   
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    #  
    print(f" : {script}")
    print(f": {summary}\n")
```
